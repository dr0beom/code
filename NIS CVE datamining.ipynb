```python
vuln_type_rule_dict = {
    'DoS': {
        'description': 'denial[- ]of[- ]service'
    },
    'Code Execution': {
        'description': '(execute.*code|execute.*file|code execution|command injection|command execution|executed|arbitrary.*code)'
    },
    'Overflow': {
        'cwe_id': 'CWE-119',
        'description': '(buffer overflow|integer overflow|stack overflow)',
    },
    'Memory Corruption': {
        'description': '((code|memory) corruption|corruption.*memory)'
    },
    'SQL Injection': {
        'cwe_id': 'CWE-89',
        'description': 'SQL (Injection|command)'
    },
    'XSS': {
        'cwe_id': 'CWE-79'
    },
    'Directory Traversal': {
        'cwe_id': 'CWE-22',
    },
    'HTTP Response Splitting': {
        'description': 'HTTP.*(request|response)[- ]splitting'
    },
    'Bypass something': {
        'description': 'Bypass'
    },
    'Gain Information': {
        'cwe_id': 'CWE-200',
        'description': 'sensitive information'
    },
    'CSRF': {
        'cwe_id': 'CWE-352',
        'description': '(CSRF|cross[- ]site[- ]request[- ]forgery )'
    },
    'File Inclusion': {
        'description': '(file inclusion|file include|inclusion.*(command|script))'
    }
}
```

```python
df['vendor_str'].value_counts()
```

```python
df_oracle=df[df['vendor_str']=='oracle']
```

```python
result = {}
for vuln_type, rule in vuln_type_rule_dict.items():
    if rule == '': continue

    tmp = pd.DataFrame()
    for column, regex in rule.items():
        filtered = df_oracle[df_oracle[column].str.contains(regex, case = False, regex = True) == True]
        tmp = pd.concat([tmp, filtered])

    result[vuln_type] = tmp.drop_duplicates(subset = 'cve_id', keep = 'first').copy()
    count = len(result[vuln_type])
    print(f'[*] {vuln_type}: {format(count, ",d")}')
```
